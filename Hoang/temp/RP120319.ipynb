```python
import pandas as pd
import xlrd

from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
# lấy danh sách skill không có khoảng trắng 
locationFile = xlrd.open_workbook("Skill review.xlsx")
g1 = locationFile.sheet_by_name("G1")
g2 = locationFile.sheet_by_name("G2")
g3 = locationFile.sheet_by_name("G3 Tech Skills")

lst_g1 = g1.col(1)[2:]
lst_g2 = g2.col(1)[2:]
lst_g3 = g3.col(0)[2:]

lst_g1 = [((item.value).upper()).split(' ') for item in lst_g1]
lst_g2 = [((item.value).upper()).split(' ') for item in lst_g2]
lst_g3 = [((item.value).upper()).split(' ') for item in lst_g3]

def split_str(_lst):
    lst = []
    for item in _lst:
        if len(item) == 1:
            lst.append(item[0])
        
    return lst

lst_skill_v1 = split_str(lst_g1) + split_str(lst_g2) + split_str(lst_g3)
```

```python
#  Load data
df1 = pd.read_excel('BasicML.xlsx')
df2 = pd.read_excel('[VN] Business Analytics & Decision Making Community.xlsx')
df3 = pd.read_excel('vietnam.data.scientists.xlsx')
```

```python
# Get list data
def parsetoList(_lst):
    lst = []
    for item in _lst:
        lst.append((item.upper()).split(' '))
    
    return lst


lst_content = parsetoList(df1['Content']) + parsetoList(df2['Content']) + parsetoList(df3['Content'])
#lst_content
```

```python
# Tính số skill
#  lst_skill_v1
#  lst_content
skill_count = [0]*len(lst_skill_v1)

for item in lst_content:
    for i in range(0, len(lst_skill_v1), 1):
        if(lst_skill_v1[i] in item):
            skill_count[i] += 1
print(skill_count)
```
